```python
import pandas as pd
import pingouin as pg
```

```python
def pw_corr(data_path, var_name):
    all_vars = pd.read_csv(data_path, index_col=0)
    pairwise = all_vars.pairwise_corr(method='spearman',padjust='holm', columns=[var_name])
    pairwise_sig = pairwise[pairwise['p-corr']<0.05]

    return pairwise_sig
```
